```python
model = YOLO('/content/drive/Shareddrives/Thesis/Data and Results/YOLOv8 Working/runs/detect/train2/weights/best.pt')
video_path = "/content/drive/Shareddrives/Thesis/Data and Results/ALL DATA/VideoProcessing/dark.mp4"
cap = cv2.VideoCapture(video_path)
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        #patches.cv2_imshow("YOLOv8 Inference", annotated_frame)

        annotated_frame = results[0].plot()
        plt.imshow(annotated_frame)
        plt.show()

         # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
